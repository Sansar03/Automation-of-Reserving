```python
lf=data.GRCODE.unique()
nf_path_for_gds="/content/drive/MyDrive/Colab resources/CsOutstandingProject/Gen_by_code not final/good_data_sets/"    #This path is used to put data sets of individual companies who have data of more than 50 rows
nf_path_for_okd="/content/drive/MyDrive/Colab resources/CsOutstandingProject/Gen_by_code not final/ok_data_sets/"      #This path is used to put data sets of individual companies who have data of more than 20 and below 50 rows
use_path="/content/drive/MyDrive/Colab resources/CsOutstandingProject/gen_for_collab/"                                 #This path is used for collecting data till 1997 and saving it so that triangles may be built over them
```

```python
j=data.groupby("GRCODE")
j
```

```python
final=[]
see=[]
for i in lf:
  j=data.groupby("GRCODE")
  p=j.get_group(i)
  name=p.loc[p.index[0],"GRNAME"]
  p=p[p["IncurLoss_C"]!=0]
  p["Cummulative_loss"]=p["IncurLoss_C"].groupby(p["AccidentYear"]).cumsum(axis=0,skipna=True)
  p["Cummulative_paid_loss"]=p["CumPaidLoss_C"]
  p["Cummulative_case_outstanding"]=(p["Cummulative_loss"]-p["Cummulative_paid_loss"])
  p["Cummulative_case_outstanding"]=p["Cummulative_case_outstanding"].groupby(p["AccidentYear"]).cumsum(axis=0,skipna=0)
  p=p[["AccidentYear","DevelopmentYear","Cummulative_loss","Cummulative_paid_loss","Cummulative_case_outstanding"]]
  p.set_index("AccidentYear")
  if (len(p)>50):
    p.to_csv(nf_path_for_gds+name+".csv",index=True)
    final.append(name)
  else:
    if(len(p)>20):
      see.append(name)
      p.to_csv(nf_path_for_okd+name+".csv")
```
